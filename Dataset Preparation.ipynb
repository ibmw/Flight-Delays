```python
import pandas as pd
import glob
```

# Trainning Dataset : Ontime 2016

```python
# Loading the data in one dataframe
list_csv = []
for filename in glob.glob('./data/*.csv'):
    # low_memory=false to avoid Mixed types error when reading csv files
    df = pd.read_csv(filename, skiprows = 0, low_memory=False)  
    list_csv.append(df)

mydf = pd.concat(list_csv)
mydf.head()
```

```python
mydf = mydf.drop("Unnamed: 25", axis=1)
cause_delay = ["CARRIER_DELAY", "WEATHER_DELAY", "NAS_DELAY", "SECURITY_DELAY", "LATE_AIRCRAFT_DELAY"]
#mydf[cause_delay] = mydf[cause_delay].fillna(0).astype("int")
mydf = mydf.drop(cause_delay, axis=1)
mydf[["DISTANCE", "FLIGHTS"]] = mydf[["DISTANCE", "FLIGHTS"]].astype("int")
```

```python
mydf = mydf.dropna(axis=0, how='any')
mydf[["DEP_DEL15", "ARR_DEL15"]] = mydf[["DEP_DEL15", "ARR_DEL15"]].astype("int")
```

```python
mydf.head()
```

```python
mydf[["ARR_TIME", "DEP_TIME"]] = mydf[["ARR_TIME", "DEP_TIME"]].astype("int")
```

```python
mydf.to_csv('delay16.csv', index=False)
```
